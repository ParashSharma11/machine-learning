```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_scores.csv")

# Display first few rows
df.head()
```

```python
# Basic information
df.info()
```

```python
# Plot relationship
plt.scatter(df["Hours_Studied"], df["Marks"])
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Hours Studied vs Marks")
plt.show()
```
